```python
import pandas as pd
import numpy as np
import os, re
from spicy import stats
import matplotlib.pyplot as plt
```

```python
files = os.listdir('.')
print(files)
```

```python
# pattern = re.compile(r'data-.*\.csv')
```

```python
files_list = []
```

```python
for file in files:
    temp_file = re.findall(r'data-.*\.csv',file)
    if temp_file is not None:
        files_list += temp_file
print(files_list)
```

```python
# files_list = re.findall(r'data-.*\.csv', for file in files)
# print(files_list)
```

```python
# print(files_list)
```

```python
for file in files_list:
    current_df = pd.read_csv(file)
    x = current_df['x']
    y = current_df['y']
    # print(x,y)

    mean_x = x.mean()
    std_x = x.std()
    min_x = x.min()
    max_x = x.max()

    mean_y = y.mean()
    std_y = y.std()
    min_y = y.min()
    max_y = y.max()

    # correlation = np.corrcoef(x,y)[0,1]
    correlation1 = stats.linregress(x,y).rvalue
    
    print(f"Summary for {file}:")
    print(f"Mean of x: {mean_x}")
    print(f"Standard Deviation of x: {std_x}")
    print(f"Range of x: ({min_x}, {max_x})")
    print(f"Mean of y: {mean_y}")
    print(f"Standard Deviation of y: {std_y}")
    print(f"Range of y: ({min_y}, {max_y})")
    # print(f"Correlation Coefficient between x and y: {correlation}")
    print(f"Correlation Coefficient between x and y: {correlation1}")
    
    print('\n')

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, alpha=0.7)
    plt.title(f'Scatter Plot of x vs y for {file}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()
    
```
